### Convolutional Neural Network (CNN)
---

Convolutional Neural Networks (CNNs) are a robust choice for breast cancer detection through image classification, particularly when applied to mammogram analysis.

Their strength lies in their ability to automatically learn hierarchical features and spatial hierarchies from 2D inputs, allowing for efficient extraction of relevant patterns indicative of breast cancer. CNNs eliminate the need for manual feature extraction, making them well-suited for image-based tasks. However, their drawbacks include a significant demand for labeled data during training, which can be a limitation in medical datasets, as well as computational intensity, requiring substantial resources. Despite these challenges, CNNs exhibit high performance, capturing complex patterns and contributing to the early detection of breast cancer.

In a practical application, a prototype can involve training a CNN model using a dataset of labeled mammogram images, where the input data comprises pixel information from these images, and the expected output is a binary classification indicating the presence or absence of breast cancer. While CNNs demonstrate remarkable capabilities, considerations should be given to their computational demands and data requirements in the context of medical image classification.



Support Vector Machines
---

Support Vector Machines (SVMs) present a robust approach for breast cancer detection through image classification, showcasing strength in handling high-dimensional data.

SVMs operate efficiently in spaces with complex data structures, making them suitable for tasks such as classifying features extracted from mammogram images. Their versatility is marked by the ability to use different kernel functions to map data into higher dimensions. However, SVMs come with their own set of challenges. While they can be powerful, they may face difficulties with large datasets, as training times may extend. Additionally, SVMs are sensitive to noise in the data and necessitate careful selection of appropriate kernel functions to handle non-linear relationships.

An example application involves using SVMs to classify features extracted from mammogram images, contributing to the identification of breast cancer. In this context, the input data would consist of these extracted features, and the expected output would be a binary classification indicating the presence or absence of breast cancer. Although SVMs offer advantages in managing high-dimensional data, their sensitivity to noise and potential challenges with large datasets should be considered in the design of breast cancer detection systems.

Transfer Learning
---

Transfer Learning emerges as a promising technique for breast cancer detection through image classification, particularly when applied to the analysis of mammogram images.

Its primary strength lies in the ability to leverage pre-trained models, such as Convolutional Neural Networks (CNNs), on large and diverse datasets like ImageNet. By fine-tuning these models for specific tasks, transfer learning mitigates the need for extensive labeled medical data, making it a pragmatic solution for limited datasets. However, this approach is not without its challenges. The assumption that knowledge from the source task is entirely relevant to the target task may not always hold, potentially leading to biased outcomes. Additionally, the transfer of biases from the source domain to the medical domain is a concern.

An example application involves fine-tuning a pre-trained CNN model on mammogram images, where the input data includes these images and possibly additional clinical data. The expected output would be a binary classification indicating the presence or absence of breast cancer. While transfer learning offers an effective compromise between model performance and data requirements, careful consideration of potential biases and domain differences is essential for its successful application in breast cancer detection systems.

```python
#pre processing the images
def img_Data(dir_path,target_size,batch,class_lst,preporcssing):
  if preporcssing:
    gen_object = ImageDataGenerator(preprocessing_function=preporcssing)
  else:
    gen_object = ImageDataGenerator()

  return(gen_object.flow_from_directory(dir_path,
                                        target_size=target_size,
                                        batch_size=batch,
                                        class_mode='sparse',
                                        classes = class_lst,
                                        shuffle=True))
```

```python
train_data_gen = img_Data("data-split/train",(224,224),500,os.listdir("data-split/train"),preprocess_input)
valid_data_gen = img_Data("data-split/test",(224,224),500,os.listdir("data-split/test"),preprocess_input)

```

```python
DATA_TRAINING_PATH = 'data-split/train'
BATCH_SIZE = 16
IMG_HEIGHT, IMG_WIDTH = 224, 224
IMG_SIZE = (IMG_HEIGHT, IMG_WIDTH, 3)

DATA_TESTING_PATH = 'data-split/test'
BATCH_SIZE = 16
IMG_HEIGHT, IMG_WIDTH = 224, 224
IMG_SIZE = (IMG_HEIGHT, IMG_WIDTH, 3)
```

```python
#splitting the data to validation and testing
train_data = tf.keras.utils.image_dataset_from_directory(
    DATA_TRAINING_PATH,
    label_mode='categorical',
    color_mode='rgb',
    validation_split=0.01,
    subset="training",
    seed=42,
    shuffle=True,
    image_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE
)

val_data = tf.keras.utils.image_dataset_from_directory(
    DATA_TESTING_PATH,
    label_mode='categorical',
    color_mode='rgb',
    validation_split=0.999,
    subset="validation",
    seed=42,
    shuffle=False,
    image_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE
)

test_labels = []
for images, labels in val_data:
    test_labels.extend(np.argmax(labels, axis=1))

test_labels = np.array(test_labels)
```

```python
class_names = train_data.class_names

# optimize
train_data = train_data.prefetch(tf.data.AUTOTUNE)
val_data = val_data.prefetch(tf.data.AUTOTUNE)

print(class_names)
```

```python
from tensorflow import keras
```

```python
#downloading resnet model
pip install resnet
```

```python
from tensorflow.keras.applications import ResNet50V2
```

```python
#frezing the model
base_model = ResNet50V2(
    include_top=False, # Exclude ImageNet classifier at the top
    weights='imagenet',
    input_shape=IMG_SIZE
    )



# Freeze the base_model
base_model.trainable = False
```

```python
#creating data augmentation to the dataset
from tensorflow.keras.layers import RandomFlip, RandomRotation, RandomTranslation, RandomContrast
data_augmentation = keras.Sequential(
    [RandomFlip('horizontal'),
     RandomRotation(factor=(-0.025, 0.025)),
     RandomTranslation(height_factor=0.1, width_factor=0.1),
     RandomContrast(factor=0.1)
     ])



# Create new model on top
inputs = keras.Input(shape=IMG_SIZE)
x = data_augmentation(inputs)

x = keras.applications.resnet_v2.preprocess_input(x)
x = base_model(x, training=False)
```

```python
#adding dropouts and activation layer
from tensorflow.keras.layers import GlobalAveragePooling2D, BatchNormalization, Dropout, Flatten, Dense
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)  # Regularize with dropout

outputs = Dense(len(class_names), activation="softmax", name="pred")(x)

resnet_model = keras.Model(inputs, outputs)



resnet_model.summary()
```

```python
#compiling the model and setting to 10 epochs to train
from tensorflow.keras.metrics import *
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy

metrics = [CategoricalAccuracy(), Precision(), Recall()]

resnet_model.compile(
    loss = CategoricalCrossentropy(),
    optimizer = Adam(learning_rate=0.001),
    metrics = metrics
    )

elst = callbacks.EarlyStopping(monitor='val_loss',patience=25,mode='min')
save_ck = callbacks.ModelCheckpoint('resnet-model-23-10-18.h5',save_best_only=True,monitor='val_loss',mode= 'min')


resnet_history = resnet_model.fit(
      train_data,
      epochs=10,
      validation_data=val_data,
      callbacks=[elst],
      verbose=1
      )


#saving the model
resnet_model.save('CWModel2.h5')


#testing it on validation dataset
result = resnet_model.evaluate(val_data)
result
```
